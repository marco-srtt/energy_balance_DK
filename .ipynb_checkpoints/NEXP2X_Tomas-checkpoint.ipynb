## Area Under the curve
In the graph above it can be seen that the orange area represents the energy supplied to the SOEC through the wind farm, and the blue area is the missing energy that is required to run the electrolyzer at its nominal power $P^{nom}$ constantly. It is therefore a key paramater indicator KPI the proportion between the total energy $E^{Nom}$ required to the cell to run at $P^{nom}$ and the actual energy supplied throught the wind farm $E^{Wind}$. 

In other words, **the Area under the curve AUC indicates a percentage of how much the SOEC is 'covered' by wind power**. For example, if at a $P^{nom} = 2$ MW the AUC is 72.5 %, this indicates at that given wind farm capacity and wind profile, the wind farm is able to supply the 72.5 % of the energy requirements, while the remaining 27.5 % is uncovered and corresponds to the blue area.

The area under the curve is evaluated as the energy $$\text{AUC} = \frac{E^{Wind}}{E^{Nom}} = \frac{\sum^{T}_{t}P^{Op}_{t}\cdot t}{\sum^{T}_{t}P^{Nom}_{t}\cdot t} $$ 
The dataframe is given in an hourly resolution, and therefore $t=1$ hour, and the sforementiond releation could further simplified:
$$\text{AUC} = \frac{\sum^{T}_{t}P^{Op}}{\sum^{T}_{t}P^{Nom}} $$


```python
SOEC_nom_power = np.linspace(0, SOEC_nom_cap, 500)        # kW
SOEC_cap_list  = np.zeros([len(SOEC_nom_power), 4])
wind_array     = np.array(df_wind['Power_output_DK_farm'].values)
SOEC_cap_list[:, 0]  = SOEC_nom_power

for count, value in enumerate(SOEC_nom_power):
    SOEC_setpoint  = np.minimum(wind_array, value)
    SOEC_cap_factor= SOEC_setpoint/value*100
    SOEC_cap_list[count, 1] = SOEC_cap_factor.mean()
    SOEC_cap_list[count, 2] = SOEC_setpoint.sum() * H2_SOEC_input
    SOEC_cap_list[count, 3] = value * H2_SOEC_input * len(wind_array)
```

```python
fig, ax1 = plt.subplots(figsize = (10, 5))

color = 'tab:red'
ax1.set_xlabel('Nominal capacity SOEC')
ax1.set_ylabel('Area under the curve %', color=color)
ax1.plot(SOEC_cap_list[:, 0], SOEC_cap_list[:, 1], color=color, label = 'Energy % from wind farm')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('H2 production', color=color)  # we already handled the x-label with ax1
ax2.plot(SOEC_cap_list[:, 0], SOEC_cap_list[:, 2], color=color,  label = '$H_{2}$ production from wind')
ax2.plot(SOEC_cap_list[:, 0], SOEC_cap_list[:, 3], color=color, linestyle = 'dashed', label = '$H_{2}$ nominal production')
ax2.tick_params(axis='y', labelcolor=color)
fig.legend(bbox_to_anchor=(0.9, 0.52))
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

# 2 - LCoH

$$OPEX_{yearly}=\text{Electricity Cost_{}}$$

For the OPEX calculation, three O&M costs are taken into account:
$$OPEX_{y}=\text{Electricity Cost}_{y} + \text{OPEX Wind farm}_{y} + \text{OPEX SOEC}_{y}$$
For the CAPEX calculation (initial investment), the following is applied:
$$CAPEX=\text{CAPEX Wind farm} + \text{CAPEX SOEC}$$
Keep in mind that since the SOEC must be replaced once its lifetime is over,
therefore there will be some capital investments for the SOEC across the project.
The income is formed by two sources: hydrogen and excess electricity production from the wind farm.
Hence, income is calcualted as:
$$\text{Hydrogen Income}_{y} = \text{Hydrogen Price} \cdot \text{SOEC Capacity} \cdot 24 \cdot 365 \cdot \text{Hydrogen Production factor SOEC}$$
$$\text{Electricity Income}_{y} = \text{Electricity Spot Price} \cdot \text{Excess wind power (power left after feeding SOEC)}_{y}$$
$$Income_{y} = \text{Hydrogen Income}_{y} + \text{Electricity Income}_{y}$$
Following on, the cash flow is calculated as (note that the CAPEX corresponding to SOEC is only included every a certain amount of years, when its lifetime concludes):
$$CashFlow_{y=0} = -CAPEX$$
$$CashFlow_{y} = -OPEX_{y} - \text{CAPEX SOEC}_{y} + Income_{y} ; 1 \leq y \leq \text{Project Lifespan}$$
Hence, the Net Present Palue (NPV) of the project is calculated as:
$$NPV = \sum_{y=0}^{\text{Project Lifespan}} \frac{CashFlow_{y}}{(1 + \text{Discount Rate})^y}$$
The internal rate of return (IRR) is the discount rate value that makes the NPV zero.
Finally, the levelized cost of hydrogen is the price at which if hydrogen was sold, then, given a certain discount rate, the NPV is zero. The thing is that it is not very clear when calcualting this levelized cost, whether the by-products sales shall be considered (because they decrease the cost when accounted for). In this cimple case the only by-product is the excess electricity produced by the wind farm.
Therefore, both levelized costs of hydrogen are calculated as follows:
$$LCoH = \frac{\sum_{y=0}^{\text{Project Lifespan}} \frac{Expenses_{y}}{(1 + \text{Discount Rate})^y}}{\sum_{y=0}^{\text{Project Lifespan}} \frac{HydrogenFlow_{y}}{(1 + \text{Discount Rate})^y}}; \forall y $$
$$LCoH_{\text{accounting for electricity sales}} = \frac{\sum_{y=0}^{\text{Project Lifespan}} \frac{Expenses_{y} - \text{Electricity Income}_{y}}{(1 + \text{Discount Rate})^y}}{\sum_{y=0}^{\text{Project Lifespan}} \frac{HydrogenFlow_{y}}{(1 + \text{Discount Rate})^y}} ;\forall y $$
$Expenses_{y}$ is obtained by summing the corresponding CAPEX and OPEX values for each year.

## OPEX

```python
Energy_production_wind_farm_yearly = df_wind['Power_output_DK_farm'].sum() * 12
Required_power          = SOEC_nom_cap * 24 * 365 - df_wind['SOEC_setpoint'].sum() * 12
Electricity_price       = 0.012 #€/kWh There is a lot of uncertainty on this value
Electricity_cost_yearly = Electricity_price * Required_power #€/year
OPEX_SOEC_yearly        = SOEC_OPEX_percentage * SOEC_specific_invest_up * SOEC_nom_cap #€/year this value must be changed when changing SOEC specific investment
OPEX_wind_farm_yearly   = OnM_fixed * wind_nom_power + OnM_var * Energy_production_wind_farm_yearly #€/year
OPEX_yearly = Electricity_cost_yearly + OPEX_wind_farm_yearly + OPEX_SOEC_yearly #€/year this is the yearly OPEX
print('The yearly OPEX is ', OPEX_yearly/1e+6, 'M€')
OPEX = [OPEX_SOEC_yearly,OPEX_wind_farm_yearly, Electricity_cost_yearly]
my_labels = ['SOEC','Onshore Wind Farm', 'Electricity']
plt.pie(OPEX,labels = my_labels,autopct = '%1.1f%%')
plt.title('OPEX distribution')
plt.axis('equal')
plt.show()
#------------------------------------CAPEX--------------------------------------------------------------------------------------------
CAPEX_wind_farm = wind_nom_power * nom_investment #€
CAPEX_SOEC      = SOEC_nom_cap * SOEC_specific_invest_up #€ this value must be changed when changing SOEC specific investment
CAPEX           = CAPEX_SOEC + CAPEX_wind_farm
print('CAPEX is',CAPEX/1e+6,' M€')
Capex = [CAPEX_SOEC, CAPEX_wind_farm]
my_labels = ['SOEC', 'Onshore Wind farm']
plt.pie(Capex,labels = my_labels,autopct = '%1.1f%%')
plt.title('Capex distribution')
plt.axis('equal')
plt.show()
#------------------------------------Income-----------------------------------------------------------------------------------------
Electricity_income_yearly  = (df_wind['Power_output_DK_farm'].sum()-df_wind['SOEC_setpoint'].sum()) * 12 * elspot_price
Hydrogen_price = 10 #€/kg   It would be wise to do a sensitivity analysis on this since the uncertainty is big
Hydrogen_production_yearly = H2_SOEC_input * SOEC_nom_cap * 24 * 365 #kg/year operating 24/7 at full capacity (when not supplied by wind farm electricity is purchased from grid)
Hydrogen_income_yearly     = Hydrogen_price * Hydrogen_production_yearly #€/year
Income_yearly              = Hydrogen_income_yearly + Electricity_income_yearly #€/year
print('The yearly income is ', Income_yearly/1e+6, ' M€')
Income = [Hydrogen_income_yearly, Electricity_income_yearly]
my_labels = ['Hydrogen', 'Electricity']
plt.pie(Income,labels = my_labels,autopct = '%1.1f%%')
plt.title('Income distribution')
plt.axis('equal')
plt.show()
#Assuming given hydrogen price
#------------------------------------CashFlow-----------------------------------------------------------------------------------------
cf = np.zeros(lifetime_wind +1) #cashflow in M€
for t in range (1, len(cf)):
    cf[t] = (- OPEX_yearly + Hydrogen_income_yearly + Electricity_income_yearly)/1e+6
cf[0] = -CAPEX/1e+6
SOEC_exact_replacement_period = SOEC_lifetime_hour_up/365/24
SOEC_replacement_period = math.floor(SOEC_exact_replacement_period)
for t in range (1+SOEC_replacement_period, len(cf),SOEC_replacement_period):
    cf[t] = -CAPEX_SOEC/1e+6 + (- OPEX_yearly + Hydrogen_income_yearly + Electricity_income_yearly)/1e+6
print('Cash Flow throughout the lifespan of the project is: ',cf,'M€')
#------------------------------------NPV--------------------------------------------------------------------------------------
discountRate    = 0.05; # Five percent per annum
npv             = npf.npv(discountRate, cf)
print('Net present value of the investment considering a ',discountRate*100,'% discount rate is: ', npv, 'M€')
#------------------------------------IRR---------------------------------------------------------------------------------------------
IRR = npf.irr(cf)
print('The IRR is ',IRR*100,'%')
#Hydrogen Price independent
#------------------------------------LCoH-----------------------------------------------------------------------------------
Expenses = np.zeros(lifetime_wind +1) #expenses in €/year
for t in range(1,len(cf)):
    Expenses[t] = -cf[t]*1e+6 + (Hydrogen_income_yearly + Electricity_income_yearly)
Expenses[0] = -cf[0]*1e+6
HydrogenFlow = np.zeros(lifetime_wind +1) #hydrogen production flows kg/h throughout the project
for t in range(1,len(cf)):
    HydrogenFlow[t] = Hydrogen_production_yearly
LCoH = npf.npv(discountRate,Expenses)/npf.npv(discountRate, HydrogenFlow)
print('The levelized cost of Hydrogen, considering a ',discountRate*100,'% discount rate, is: ',LCoH,'€/kg    #Here we are not accounting for wind farm excess electricity production or other sources of income')
ExpensesPlusElectricityIncome = np.zeros(lifetime_wind +1) #expenses plus electricity income in €/year
for t in range(1,len(cf)):
    ExpensesPlusElectricityIncome[t] = -cf[t]*1e+6 + (Hydrogen_income_yearly)
ExpensesPlusElectricityIncome[0] = -cf[0]*1e+6
LCoH2 = npf.npv(discountRate,ExpensesPlusElectricityIncome)/npf.npv(discountRate, HydrogenFlow)
print('The levelized cost of Hydrogen, considering a ',discountRate*100,'% discount rate, is: ',LCoH2,'€/kg    #Here we are accounting for wind farm excess electricity production')
```

```python
fig, ax = plt.subplots(1, 2, figsize = (15,5), sharex = True, sharey = True)
ax[0].bar(range(len(cf)), cf)
ax[0].set_title('Cashflow')
ax[0].set_xlabel('Years')
ax[0].set_ylabel('€')
ax[1].bar(range(len(cf)), np.cumsum(cf))
ax[1].set_title('NPV')
ax[1].set_xlabel('Years')
ax[1].set_ylabel('€')
plt.show()
```

# 3 - Sensitivity analysis on SOEC Specific investment

```python
number_of_points=100
MaxValue = 6000 #€/kg
SOEC_specific_invest   = np.zeros(number_of_points+1)
SOEC_specific_invest[0]= 1000 #€/kg
step = (MaxValue - SOEC_specific_invest[0])/(number_of_points)
for t in range(1,number_of_points+1):
    SOEC_specific_invest[t] = SOEC_specific_invest[t-1] + step
#print(SOEC_specific_invest)
#Run all together since only the following changes in the previous code:
#------------------------------------OPEX---------------------------------------------------------------------------------------------
OPEX_SOEC_yearly        = SOEC_OPEX_percentage * SOEC_specific_invest * SOEC_nom_cap #€/year
OPEX_yearly = Electricity_cost_yearly + OPEX_wind_farm_yearly + OPEX_SOEC_yearly #€/year
#------------------------------------CAPEX--------------------------------------------------------------------------------------------
CAPEX_SOEC      = SOEC_nom_cap * SOEC_specific_invest #€
CAPEX           = CAPEX_SOEC + CAPEX_wind_farm #€
#Assuming given hydrogen price
#------------------------------------CashFlow-----------------------------------------------------------------------------------------
cf = np.zeros((number_of_points +1,lifetime_wind +1)) #cashflow in M€
for p in range(0,number_of_points+1):
    for t in range (1, lifetime_wind +1):
        cf[p][t] = (- OPEX_yearly[p] + Hydrogen_income_yearly + Electricity_income_yearly)/1e+6
    cf[p][0] = -CAPEX[p]/1e+6
    for t in range (1+SOEC_replacement_period, lifetime_wind +1,SOEC_replacement_period):
        cf[p][t] = -CAPEX_SOEC[p]/1e+6 + (- OPEX_yearly[p] + Hydrogen_income_yearly + Electricity_income_yearly)/1e+6
#print('Cash Flow throughout the lifespan of the project is: ',cf,'M€')
#------------------------------------NPV--------------------------------------------------------------------------------------
npv             = np.zeros(number_of_points+1)
discountRate    = 0.05; # Five percent per annum
for p in range(0,number_of_points+1):
    npv[p]             = npf.npv(discountRate, cf[p][:])
#print('Net present value of the investment considering a ',discountRate*100,'% discount rate is: ', npv, 'M€')
#------------------------------------IRR---------------------------------------------------------------------------------------------
IRR = np.zeros(number_of_points+1)
for p in range(0,number_of_points+1):
    IRR[p] = npf.irr(cf[p][:])
#print('The IRR is: ',IRR*100,'%')
#Hydrogen Price independent
#------------------------------------LCoH-----------------------------------------------------------------------------------
Expenses = np.zeros((number_of_points +1,lifetime_wind +1)) #expenses in €/year
LCoH = np.zeros(number_of_points +1) #levelized cost of hydrogen in €/kg
LCoH2 = np.zeros(number_of_points +1) #levelized cost of hydrogen in €/kg accounting for electricity sales
for p in range(0,number_of_points+1):
    for t in range(1,lifetime_wind +1):
        Expenses[p][t] = -cf[p][t]*1e+6 + (Hydrogen_income_yearly + Electricity_income_yearly)
    Expenses[p][0] = -cf[p][0]*1e+6
HydrogenFlow = np.zeros((number_of_points +1,lifetime_wind +1)) #hydrogen production flows kg/h throughout the project
for p in range(0,number_of_points+1):
    for t in range(1,lifetime_wind +1):
        HydrogenFlow[p][t] = Hydrogen_production_yearly
    LCoH[p] = npf.npv(discountRate,Expenses[p][:])/npf.npv(discountRate, HydrogenFlow[p][:])
#print('The levelized cost of Hydrogen, considering a ',discountRate*100,'% discount rate, is: ',LCoH,'€/kg    #Here we are not accounting for wind farm excess electricity production or other sources of income')
ExpensesPlusElectricityIncome = np.zeros((number_of_points +1,lifetime_wind +1)) #expenses plus electricity income in €/year
for p in range(0,number_of_points+1):
    for t in range(1,lifetime_wind +1):
        ExpensesPlusElectricityIncome[p][t] = -cf[p][t]*1e+6 + (Hydrogen_income_yearly)
    ExpensesPlusElectricityIncome[p][0] = -cf[p][0]*1e+6
    LCoH2[p] = npf.npv(discountRate,ExpensesPlusElectricityIncome[p][:])/npf.npv(discountRate, HydrogenFlow[p][:])
#print('The levelized cost of Hydrogen, considering a ',discountRate*100,'% discount rate, is: ',LCoH2,'€/kg    #Here we are accounting for wind farm excess electricity production')
#-------------------------------------Plots------------------------------------------------------------------------------------------
print('The following plot shows the LCoH that accounts for the sales of excess electricity production from the wind farm, considering a hydrogen price of',Hydrogen_price,'€/kg and a SOEC nomincal capacity of',SOEC_nom_cap,'kW:')
fig, ax1 = plt.subplots(figsize = (10, 5))
color = 'tab:red'
ax1.set_xlabel('SOEC Specific Investment [€/kW]')
ax1.set_ylabel('LCoH [€/kg]', color=color)
ax1.plot(SOEC_specific_invest[:], LCoH2[:], color=color, label = 'Levelized cost of Hydrogen (Minimum price for Hydrogen)') #minimum price of hydrogen in order to have an IRR greater than zero for a given SOEC specific Investment
ax1.tick_params(axis='y', labelcolor=color)
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:blue'
ax2.set_ylabel('IRR %', color=color)  # we already handled the x-label with ax1
ax2.plot(SOEC_specific_invest[:], IRR*100, color=color,  label = 'Internal rate of return with hydrogen price of ' +str(Hydrogen_price) +' €/kg')
ax2.tick_params(axis='y', labelcolor=color)
plt.axvline(x=SOEC_specific_invest_low, linestyle = "dashed")
plt.axvline(x=SOEC_specific_invest_up, linestyle = "dashed")
fig.legend(bbox_to_anchor=(0.7, 0.25))
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

# 4 - Sensitivity analysis on SOEC nominal capacity

```python
SOEC_nom_capa=1000 #kW
Hydrogen_production_yearly = H2_SOEC_input * SOEC_nom_capa * 24 * 365 #kg/year operating 24/7 at full capacity (when not supplied by wind farm electricity is purchased from grid)
Hydrogen_income_yearly     = Hydrogen_price * Hydrogen_production_yearly #€/year
df_wind['SOEC_setpoint'] = np.minimum(df_wind['Power_output_DK_farm'], SOEC_nom_capa)
Required_power          = SOEC_nom_capa * 24 * 365 - df_wind['SOEC_setpoint'].sum() * 12
Electricity_price       = 0.012 #€/kWh There is a lot of uncertainty on this value
Electricity_cost_yearly = Electricity_price * Required_power #€/year
number_of_points=100
MaxValue = 6000 #€/kg
SOEC_specific_invest   = np.zeros(number_of_points+1)
SOEC_specific_invest[0]= 1000 #€/kg
step = (MaxValue - SOEC_specific_invest[0])/(number_of_points)
for t in range(1,number_of_points+1):
    SOEC_specific_invest[t] = SOEC_specific_invest[t-1] + step
#print(SOEC_specific_invest)
#Run all together since only the following changes in the previous code:
#------------------------------------OPEX---------------------------------------------------------------------------------------------
OPEX_SOEC_yearly        = SOEC_OPEX_percentage * SOEC_specific_invest * SOEC_nom_capa #€/year
OPEX_yearly = Electricity_cost_yearly + OPEX_wind_farm_yearly + OPEX_SOEC_yearly #€/year
#------------------------------------CAPEX--------------------------------------------------------------------------------------------
CAPEX_SOEC      = SOEC_nom_capa * SOEC_specific_invest #€
CAPEX           = CAPEX_SOEC + CAPEX_wind_farm #€
#Assuming given hydrogen price
#------------------------------------CashFlow-----------------------------------------------------------------------------------------
cf = np.zeros((number_of_points +1,lifetime_wind +1)) #cashflow in M€
for p in range(0,number_of_points+1):
    for t in range (1, lifetime_wind +1):
        cf[p][t] = (- OPEX_yearly[p] + Hydrogen_income_yearly + Electricity_income_yearly)/1e+6
    cf[p][0] = -CAPEX[p]/1e+6
    for t in range (1+SOEC_replacement_period, lifetime_wind +1,SOEC_replacement_period):
        cf[p][t] = -CAPEX_SOEC[p]/1e+6 + (- OPEX_yearly[p] + Hydrogen_income_yearly + Electricity_income_yearly)/1e+6
#print('Cash Flow throughout the lifespan of the project is: ',cf,'M€')
#------------------------------------NPV--------------------------------------------------------------------------------------
npv             = np.zeros(number_of_points+1)
discountRate    = 0.05; # Five percent per annum
for p in range(0,number_of_points+1):
    npv[p]             = npf.npv(discountRate, cf[p][:])
#print('Net present value of the investment considering a ',discountRate*100,'% discount rate is: ', npv, 'M€')
#------------------------------------IRR---------------------------------------------------------------------------------------------
IRRa = np.zeros(number_of_points+1)
for p in range(0,number_of_points+1):
    IRRa[p] = npf.irr(cf[p][:])
#print('The IRR is: ',IRR*100,'%')
#Hydrogen Price independent
#------------------------------------LCoH-----------------------------------------------------------------------------------
Expenses = np.zeros((number_of_points +1,lifetime_wind +1)) #expenses in €/year
LCoH = np.zeros(number_of_points +1) #levelized cost of hydrogen in €/kg
LCoH2a = np.zeros(number_of_points +1) #levelized cost of hydrogen in €/kg accounting for electricity sales
for p in range(0,number_of_points+1):
    for t in range(1,lifetime_wind +1):
        Expenses[p][t] = -cf[p][t]*1e+6 + (Hydrogen_income_yearly + Electricity_income_yearly)
    Expenses[p][0] = -cf[p][0]*1e+6
HydrogenFlow = np.zeros((number_of_points +1,lifetime_wind +1)) #hydrogen production flows kg/h throughout the project
for p in range(0,number_of_points+1):
    for t in range(1,lifetime_wind +1):
        HydrogenFlow[p][t] = Hydrogen_production_yearly
    LCoH[p] = npf.npv(discountRate,Expenses[p][:])/npf.npv(discountRate, HydrogenFlow[p][:])
#print('The levelized cost of Hydrogen, considering a ',discountRate*100,'% discount rate, is: ',LCoH,'€/kg    #Here we are not accounting for wind farm excess electricity production or other sources of income')
ExpensesPlusElectricityIncome = np.zeros((number_of_points +1,lifetime_wind +1)) #expenses plus electricity income in €/year
for p in range(0,number_of_points+1):
    for t in range(1,lifetime_wind +1):
        ExpensesPlusElectricityIncome[p][t] = -cf[p][t]*1e+6 + (Hydrogen_income_yearly)
    ExpensesPlusElectricityIncome[p][0] = -cf[p][0]*1e+6
    LCoH2a[p] = npf.npv(discountRate,ExpensesPlusElectricityIncome[p][:])/npf.npv(discountRate, HydrogenFlow[p][:])
#print('The levelized cost of Hydrogen, considering a ',discountRate*100,'% discount rate, is: ',LCoH2a,'€/kg    #Here we are accounting for wind farm excess electricity production')

SOEC_nom_capb=2000 #kW
Hydrogen_production_yearly = H2_SOEC_input * SOEC_nom_capb * 24 * 365 #kg/year operating 24/7 at full capacity (when not supplied by wind farm electricity is purchased from grid)
Hydrogen_income_yearly     = Hydrogen_price * Hydrogen_production_yearly #€/year
df_wind['SOEC_setpoint'] = np.minimum(df_wind['Power_output_DK_farm'], SOEC_nom_capb)
Required_power          = SOEC_nom_capb * 24 * 365 - df_wind['SOEC_setpoint'].sum() * 12
Electricity_price       = 0.012 #€/kWh There is a lot of uncertainty on this value
Electricity_cost_yearly = Electricity_price * Required_power #€/year
number_of_points=100
MaxValue = 6000 #€/kg
SOEC_specific_invest   = np.zeros(number_of_points+1)
SOEC_specific_invest[0]= 1000 #€/kg
step = (MaxValue - SOEC_specific_invest[0])/(number_of_points)
for t in range(1,number_of_points+1):
    SOEC_specific_invest[t] = SOEC_specific_invest[t-1] + step
#print(SOEC_specific_invest)
#Run all together since only the following changes in the previous code:
#------------------------------------OPEX---------------------------------------------------------------------------------------------
OPEX_SOEC_yearly        = SOEC_OPEX_percentage * SOEC_specific_invest * SOEC_nom_capb #€/year
OPEX_yearly = Electricity_cost_yearly + OPEX_wind_farm_yearly + OPEX_SOEC_yearly #€/year
#------------------------------------CAPEX--------------------------------------------------------------------------------------------
CAPEX_SOEC      = SOEC_nom_capb * SOEC_specific_invest #€
CAPEX           = CAPEX_SOEC + CAPEX_wind_farm #€
#Assuming given hydrogen price
#------------------------------------CashFlow-----------------------------------------------------------------------------------------
cf = np.zeros((number_of_points +1,lifetime_wind +1)) #cashflow in M€
for p in range(0,number_of_points+1):
    for t in range (1, lifetime_wind +1):
        cf[p][t] = (- OPEX_yearly[p] + Hydrogen_income_yearly + Electricity_income_yearly)/1e+6
    cf[p][0] = -CAPEX[p]/1e+6
    for t in range (1+SOEC_replacement_period, lifetime_wind +1,SOEC_replacement_period):
        cf[p][t] = -CAPEX_SOEC[p]/1e+6 + (- OPEX_yearly[p] + Hydrogen_income_yearly + Electricity_income_yearly)/1e+6
#print('Cash Flow throughout the lifespan of the project is: ',cf,'M€')
#------------------------------------NPV--------------------------------------------------------------------------------------
npv             = np.zeros(number_of_points+1)
discountRate    = 0.05; # Five percent per annum
for p in range(0,number_of_points+1):
    npv[p]             = npf.npv(discountRate, cf[p][:])
#print('Net present value of the investment considering a ',discountRate*100,'% discount rate is: ', npv, 'M€')
#------------------------------------IRR---------------------------------------------------------------------------------------------
IRRb = np.zeros(number_of_points+1)
for p in range(0,number_of_points+1):
    IRRb[p] = npf.irr(cf[p][:])
#print('The IRR is: ',IRR*100,'%')
#Hydrogen Price independent
#------------------------------------LCoH-----------------------------------------------------------------------------------
Expenses = np.zeros((number_of_points +1,lifetime_wind +1)) #expenses in €/year
LCoH = np.zeros(number_of_points +1) #levelized cost of hydrogen in €/kg
LCoH2b = np.zeros(number_of_points +1) #levelized cost of hydrogen in €/kg accounting for electricity sales
for p in range(0,number_of_points+1):
    for t in range(1,lifetime_wind +1):
        Expenses[p][t] = -cf[p][t]*1e+6 + (Hydrogen_income_yearly + Electricity_income_yearly)
    Expenses[p][0] = -cf[p][0]*1e+6
HydrogenFlow = np.zeros((number_of_points +1,lifetime_wind +1)) #hydrogen production flows kg/h throughout the project
for p in range(0,number_of_points+1):
    for t in range(1,lifetime_wind +1):
        HydrogenFlow[p][t] = Hydrogen_production_yearly
    LCoH[p] = npf.npv(discountRate,Expenses[p][:])/npf.npv(discountRate, HydrogenFlow[p][:])
#print('The levelized cost of Hydrogen, considering a ',discountRate*100,'% discount rate, is: ',LCoH,'€/kg    #Here we are not accounting for wind farm excess electricity production or other sources of income')
ExpensesPlusElectricityIncome = np.zeros((number_of_points +1,lifetime_wind +1)) #expenses plus electricity income in €/year
for p in range(0,number_of_points+1):
    for t in range(1,lifetime_wind +1):
        ExpensesPlusElectricityIncome[p][t] = -cf[p][t]*1e+6 + (Hydrogen_income_yearly)
    ExpensesPlusElectricityIncome[p][0] = -cf[p][0]*1e+6
    LCoH2b[p] = npf.npv(discountRate,ExpensesPlusElectricityIncome[p][:])/npf.npv(discountRate, HydrogenFlow[p][:])
#print('The levelized cost of Hydrogen, considering a ',discountRate*100,'% discount rate, is: ',LCoH2a,'€/kg    #Here we are accounting for wind farm excess electricity production')
SOEC_nom_capc=3000 #kW
Hydrogen_production_yearly = H2_SOEC_input * SOEC_nom_capc * 24 * 365 #kg/year operating 24/7 at full capacity (when not supplied by wind farm electricity is purchased from grid)
Hydrogen_income_yearly     = Hydrogen_price * Hydrogen_production_yearly #€/year
df_wind['SOEC_setpoint'] = np.minimum(df_wind['Power_output_DK_farm'], SOEC_nom_capc)
Required_power          = SOEC_nom_capc * 24 * 365 - df_wind['SOEC_setpoint'].sum() * 12
Electricity_price       = 0.012 #€/kWh There is a lot of uncertainty on this value
Electricity_cost_yearly = Electricity_price * Required_power #€/year
number_of_points=100
MaxValue = 6000 #€/kg
SOEC_specific_invest   = np.zeros(number_of_points+1)
SOEC_specific_invest[0]= 1000 #€/kg
step = (MaxValue - SOEC_specific_invest[0])/(number_of_points)
for t in range(1,number_of_points+1):
    SOEC_specific_invest[t] = SOEC_specific_invest[t-1] + step
#print(SOEC_specific_invest)
#Run all together since only the following changes in the previous code:
#------------------------------------OPEX---------------------------------------------------------------------------------------------
OPEX_SOEC_yearly        = SOEC_OPEX_percentage * SOEC_specific_invest * SOEC_nom_capc #€/year
OPEX_yearly = Electricity_cost_yearly + OPEX_wind_farm_yearly + OPEX_SOEC_yearly #€/year
#------------------------------------CAPEX--------------------------------------------------------------------------------------------
CAPEX_SOEC      = SOEC_nom_capc * SOEC_specific_invest #€
CAPEX           = CAPEX_SOEC + CAPEX_wind_farm #€
#Assuming given hydrogen price
#------------------------------------CashFlow-----------------------------------------------------------------------------------------
cf = np.zeros((number_of_points +1,lifetime_wind +1)) #cashflow in M€
for p in range(0,number_of_points+1):
    for t in range (1, lifetime_wind +1):
        cf[p][t] = (- OPEX_yearly[p] + Hydrogen_income_yearly + Electricity_income_yearly)/1e+6
    cf[p][0] = -CAPEX[p]/1e+6
    for t in range (1+SOEC_replacement_period, lifetime_wind +1,SOEC_replacement_period):
        cf[p][t] = -CAPEX_SOEC[p]/1e+6 + (- OPEX_yearly[p] + Hydrogen_income_yearly + Electricity_income_yearly)/1e+6
#print('Cash Flow throughout the lifespan of the project is: ',cf,'M€')
#------------------------------------NPV--------------------------------------------------------------------------------------
npv             = np.zeros(number_of_points+1)
discountRate    = 0.05; # Five percent per annum
for p in range(0,number_of_points+1):
    npv[p]             = npf.npv(discountRate, cf[p][:])
#print('Net present value of the investment considering a ',discountRate*100,'% discount rate is: ', npv, 'M€')
#------------------------------------IRR---------------------------------------------------------------------------------------------
IRRc = np.zeros(number_of_points+1)
for p in range(0,number_of_points+1):
    IRRc[p] = npf.irr(cf[p][:])
#print('The IRR is: ',IRR*100,'%')
#Hydrogen Price independent
#------------------------------------LCoH-----------------------------------------------------------------------------------
Expenses = np.zeros((number_of_points +1,lifetime_wind +1)) #expenses in €/year
LCoH = np.zeros(number_of_points +1) #levelized cost of hydrogen in €/kg
LCoH2c = np.zeros(number_of_points +1) #levelized cost of hydrogen in €/kg accounting for electricity sales
for p in range(0,number_of_points+1):
    for t in range(1,lifetime_wind +1):
        Expenses[p][t] = -cf[p][t]*1e+6 + (Hydrogen_income_yearly + Electricity_income_yearly)
    Expenses[p][0] = -cf[p][0]*1e+6
HydrogenFlow = np.zeros((number_of_points +1,lifetime_wind +1)) #hydrogen production flows kg/h throughout the project
for p in range(0,number_of_points+1):
    for t in range(1,lifetime_wind +1):
        HydrogenFlow[p][t] = Hydrogen_production_yearly
    LCoH[p] = npf.npv(discountRate,Expenses[p][:])/npf.npv(discountRate, HydrogenFlow[p][:])
#print('The levelized cost of Hydrogen, considering a ',discountRate*100,'% discount rate, is: ',LCoH,'€/kg    #Here we are not accounting for wind farm excess electricity production or other sources of income')
ExpensesPlusElectricityIncome = np.zeros((number_of_points +1,lifetime_wind +1)) #expenses plus electricity income in €/year
for p in range(0,number_of_points+1):
    for t in range(1,lifetime_wind +1):
        ExpensesPlusElectricityIncome[p][t] = -cf[p][t]*1e+6 + (Hydrogen_income_yearly)
    ExpensesPlusElectricityIncome[p][0] = -cf[p][0]*1e+6
    LCoH2c[p] = npf.npv(discountRate,ExpensesPlusElectricityIncome[p][:])/npf.npv(discountRate, HydrogenFlow[p][:])
#print('The levelized cost of Hydrogen, considering a ',discountRate*100,'% discount rate, is: ',LCoH2a,'€/kg    #Here we are accounting for wind farm excess electricity production')

SOEC_nom_capd=5000 #kW
Hydrogen_production_yearly = H2_SOEC_input * SOEC_nom_capd * 24 * 365 #kg/year operating 24/7 at full capacity (when not supplied by wind farm electricity is purchased from grid)
Hydrogen_income_yearly     = Hydrogen_price * Hydrogen_production_yearly #€/year
df_wind['SOEC_setpoint'] = np.minimum(df_wind['Power_output_DK_farm'], SOEC_nom_capd)
Required_power          = SOEC_nom_capd * 24 * 365 - df_wind['SOEC_setpoint'].sum() * 12
Electricity_price       = 0.012 #€/kWh There is a lot of uncertainty on this value
Electricity_cost_yearly = Electricity_price * Required_power #€/year
number_of_points=100
MaxValue = 6000 #€/kg
SOEC_specific_invest   = np.zeros(number_of_points+1)
SOEC_specific_invest[0]= 1000 #€/kg
step = (MaxValue - SOEC_specific_invest[0])/(number_of_points)
for t in range(1,number_of_points+1):
    SOEC_specific_invest[t] = SOEC_specific_invest[t-1] + step
#print(SOEC_specific_invest)
#Run all together since only the following changes in the previous code:
#------------------------------------OPEX---------------------------------------------------------------------------------------------
OPEX_SOEC_yearly        = SOEC_OPEX_percentage * SOEC_specific_invest * SOEC_nom_capd #€/year
OPEX_yearly = Electricity_cost_yearly + OPEX_wind_farm_yearly + OPEX_SOEC_yearly #€/year
#------------------------------------CAPEX--------------------------------------------------------------------------------------------
CAPEX_SOEC      = SOEC_nom_capd * SOEC_specific_invest #€
CAPEX           = CAPEX_SOEC + CAPEX_wind_farm #€
#Assuming given hydrogen price
#------------------------------------CashFlow-----------------------------------------------------------------------------------------
cf = np.zeros((number_of_points +1,lifetime_wind +1)) #cashflow in M€
for p in range(0,number_of_points+1):
    for t in range (1, lifetime_wind +1):
        cf[p][t] = (- OPEX_yearly[p] + Hydrogen_income_yearly + Electricity_income_yearly)/1e+6
    cf[p][0] = -CAPEX[p]/1e+6
    for t in range (1+SOEC_replacement_period, lifetime_wind +1,SOEC_replacement_period):
        cf[p][t] = -CAPEX_SOEC[p]/1e+6 + (- OPEX_yearly[p] + Hydrogen_income_yearly + Electricity_income_yearly)/1e+6
#print('Cash Flow throughout the lifespan of the project is: ',cf,'M€')
#------------------------------------NPV--------------------------------------------------------------------------------------
npv             = np.zeros(number_of_points+1)
discountRate    = 0.05; # Five percent per annum
for p in range(0,number_of_points+1):
    npv[p]             = npf.npv(discountRate, cf[p][:])
#print('Net present value of the investment considering a ',discountRate*100,'% discount rate is: ', npv, 'M€')
#------------------------------------IRR---------------------------------------------------------------------------------------------
IRRd = np.zeros(number_of_points+1)
for p in range(0,number_of_points+1):
    IRRd[p] = npf.irr(cf[p][:])
#print('The IRR is: ',IRR*100,'%')
#Hydrogen Price independent
#------------------------------------LCoH-----------------------------------------------------------------------------------
Expenses = np.zeros((number_of_points +1,lifetime_wind +1)) #expenses in €/year
LCoH = np.zeros(number_of_points +1) #levelized cost of hydrogen in €/kg
LCoH2d = np.zeros(number_of_points +1) #levelized cost of hydrogen in €/kg accounting for electricity sales
for p in range(0,number_of_points+1):
    for t in range(1,lifetime_wind +1):
        Expenses[p][t] = -cf[p][t]*1e+6 + (Hydrogen_income_yearly + Electricity_income_yearly)
    Expenses[p][0] = -cf[p][0]*1e+6
HydrogenFlow = np.zeros((number_of_points +1,lifetime_wind +1)) #hydrogen production flows kg/h throughout the project
for p in range(0,number_of_points+1):
    for t in range(1,lifetime_wind +1):
        HydrogenFlow[p][t] = Hydrogen_production_yearly
    LCoH[p] = npf.npv(discountRate,Expenses[p][:])/npf.npv(discountRate, HydrogenFlow[p][:])
#print('The levelized cost of Hydrogen, considering a ',discountRate*100,'% discount rate, is: ',LCoH,'€/kg    #Here we are not accounting for wind farm excess electricity production or other sources of income')
ExpensesPlusElectricityIncome = np.zeros((number_of_points +1,lifetime_wind +1)) #expenses plus electricity income in €/year
for p in range(0,number_of_points+1):
    for t in range(1,lifetime_wind +1):
        ExpensesPlusElectricityIncome[p][t] = -cf[p][t]*1e+6 + (Hydrogen_income_yearly)
    ExpensesPlusElectricityIncome[p][0] = -cf[p][0]*1e+6
    LCoH2d[p] = npf.npv(discountRate,ExpensesPlusElectricityIncome[p][:])/npf.npv(discountRate, HydrogenFlow[p][:])
#print('The levelized cost of Hydrogen, considering a ',discountRate*100,'% discount rate, is: ',LCoH2a,'€/kg    #Here we are accounting for wind farm excess electricity production')
#-------------------------------------Plots------------------------------------------------------------------------------------------
print('The following plot shows the LCoH that accounts for the sales of excess electricity production from the wind farm, considering a hydrogen price of',Hydrogen_price,'€/kg:')
fig, ax1 = plt.subplots(figsize = (10, 5))
color = 'tab:red'
ax1.set_xlabel('SOEC Specific Investment [€/kW]')
ax1.set_ylabel('LCoH [€/kg]', color=color)
ax1.plot(SOEC_specific_invest[:], LCoH2a[:], color=color, label = 'Levelized cost of Hydrogen with SOEC capacity of ' +str(SOEC_nom_capa)+'kW') #minimum price of hydrogen in order to have an IRR greater than zero for a given SOEC specific Investment
ax1.plot(SOEC_specific_invest[:], LCoH2b[:], color=color,linestyle = 'dashed', label = 'Levelized cost of Hydrogen with SOEC capacity of ' +str(SOEC_nom_capb)+'kW') #minimum price of hydrogen in order to have an IRR greater than zero for a given SOEC specific Investment
ax1.plot(SOEC_specific_invest[:], LCoH2c[:], color=color,linestyle = 'dotted', label = 'Levelized cost of Hydrogen with SOEC capacity of ' +str(SOEC_nom_capc)+'kW') #minimum price of hydrogen in order to have an IRR greater than zero for a given SOEC specific Investment
ax1.plot(SOEC_specific_invest[:], LCoH2d[:], color=color,linestyle = 'dashdot', label = 'Levelized cost of Hydrogen with SOEC capacity of ' +str(SOEC_nom_capd)+'kW') #minimum price of hydrogen in order to have an IRR greater than zero for a given SOEC specific Investment
ax1.tick_params(axis='y', labelcolor=color)
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:blue'
ax2.set_ylabel('IRR %', color=color)  # we already handled the x-label with ax1
ax2.plot(SOEC_specific_invest[:], IRRa*100, color=color,  label = 'Internal rate of return with SOEC capacity of ' +str(SOEC_nom_capa) +' kW')
ax2.plot(SOEC_specific_invest[:], IRRb*100, color=color,linestyle = 'dashed',  label = 'Internal rate of return with SOEC capacity of ' +str(SOEC_nom_capb) +' kW')
ax2.plot(SOEC_specific_invest[:], IRRc*100, color=color,linestyle = 'dotted',  label = 'Internal rate of return with SOEC capacity of ' +str(SOEC_nom_capc) +' kW')
ax2.plot(SOEC_specific_invest[:], IRRd*100, color=color,linestyle = 'dashdot',  label = 'Internal rate of return with SOEC capacity of ' +str(SOEC_nom_capd) +' kW')
ax2.tick_params(axis='y', labelcolor=color)
fig.legend(bbox_to_anchor=(0.7, 0))
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
