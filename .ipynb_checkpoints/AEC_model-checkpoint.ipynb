# NEXP2X - Alkaline electrolyzer model

## Package import

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import handcalcs.render
from periodictable import H, O
%matplotlib inline
```

## Electrolyzer setting

```python
#%% CONSTANTS
n = 2          # Electrons exchanged
F = 96485      # Farady constant
delta_H_HHV = 285.8 # kJ/mol
delta_H_LHV = 241.8 # kJ/mol

delta_G_HHV = 237.1 # kJ/mol    25 C, 1 bar
delta_G_LHV = 228.6 # kJ/mol    25 C, 1 bar

#%% ELECTROLYZER SETTINGS

K   = 273.15        # Kelvin conversion
p   = 7             # Pressure [bar]
N   = 1000          # Number of stack
T   = 80            # Temperature C
i   = 5000          # Current Density A/m2
A   = 0.1           # Area electrode m2

MW_H2O = H.mass*2+O.mass
V_rev_25 = delta_G_HHV*1000/(n*F)      # Reversible voltage
V_tn   = delta_H_HHV*1000/(n*F)      # Thermoneautral voltage    
V_rev  = 1.5184 - 1.5421*10**-3* (T+K) + (9.523*10**-5)*(T+K)*np.log((T+K)) +(9.84*10**-8)*(T+K)**2
```

## Constats for polynomial regression

```python
# POLARIZATION CURVE
r_1 = 4.451530E-05
r_2 = 6.888740E-09
d_1 = -3.129960E-06
d_2 = 4.471370E-07
s   = 3.382400E-01
t_1 = -1.539000E-02
t_2 = 2.001810
t_3 = 1.524178E+01

# FARADAIC EFFICIENCY
f_11 = 4.786457E+05
f_12 = -2.953150E+03
f_21 = 1.039600E+00
f_22 = -1.040000E-03

# GAS PURITY 
C_1 = 9.901000E-02
C_2 = -2.070000E-03
C_3 = 1.310640E-05
C_4 = -8.483000E-02
C_5 = 1.790000E-03
C_6 = -1.133900E-05
C_7 = 1.481450E+03
C_8 = -2.360345E+01
C_9 = -2.577400E-01
E_1 = 3.714170E+00
E_2 = -9.306300E-01
E_3 = 5.817000E-02
E_4 = -3.720680E+00
E_5 = 9.321900E-01
E_6 = -5.826000E-02
E_7 = -1.838215E+01
E_8 = 5.873160E+00
E_9 = -4.642500E-01
```

## Voltage - Current curve
The voltage required from an electrolyzer is given by the reversible voltage plus the overpotentials expressed as follows:
$$
V_{\text {cell }}=V_{r e v}+\left(\hat{\eta}_{\text {cat }}+\widehat{\eta}_{a n}+\widehat{\eta}_{o h m}+\widehat{\eta}_{\text {conc }}\right)
$$

The voltage term could be also approximated by a polynomial :

$$
V_{\text {cell }}=V_{rev}+\left[\left(r_{1}+d_{1}\right)+r_{2} \cdot T+d_{2} \cdot p\right] i+s \cdot \log \left[\left(t_{1}+\frac{t_{2}}{T}+\frac{t_{3}}{T^{2}}\right) i+1\right]
$$

```python
V_cell = V_rev + ((r_1 + d_1) + r_2 * T + d_2 * p) * i + s * np.log10((t_1 + t_2/T + t_3 / T**2)*i + 1)
P_cell = V_cell * i * A
P_stack = P_cell * N

print('The cell voltage is ' + str(round(V_cell, 2)) + ' V and the current ' + str(i)+' A/m2, with an area of '+ str(A) + ' m2')
print('The Power of a single cell is ' + str(round(P_cell, 2)) + ' W')
print('The Power for '+ str(N) + ' stacks is ' + str(round(P_stack/1000000, 2)) + ' MW')
```

```python
x = np.linspace(0, 10000, 1000)
rev_plot = np.ones(len(x))*V_rev
ohm_plot = ((r_1 + d_1) + r_2 * T + d_2 * p) * x
act_plot = s * np.log10((t_1 + t_2/T + t_3 / T**2)*x + 1)
plt.figure(figsize=(7,5))
plt.plot(x, rev_plot)
plt.plot(x, rev_plot + ohm_plot)
plt.plot(x, rev_plot + ohm_plot + act_plot)
plt.xlabel('Current density [A/m$^2$]')
plt.ylabel('Cell voltage [V]')
plt.title('Polarization curve AEC, '+str(T)+' C')
plt.fill_between(x, rev_plot, rev_plot + ohm_plot , color='tab:orange', alpha=0.5)
plt.fill_between(x, rev_plot + ohm_plot , rev_plot + ohm_plot + act_plot , color='tab:green', alpha=0.5)

box_style=dict(boxstyle='round', facecolor='wheat', alpha=0.5)
plt.text(6000, 1.3, "Ohmic overpotential", {'color':'black','size':10},bbox=box_style)

box_style=dict(boxstyle='round', facecolor='wheat', alpha=0.5)
plt.text(6000, 1.8, "Activation overpotential", {'color':'black','size':10},bbox=box_style)
```

## Faradaic efficiency
The faradaic efficiency is defined as the real molar production via electrolysis over the theoretical expected production:
$$
\eta_{F}=\frac{n_{H 2, p r o d}}{n_{H 2, t h}}
$$

It can be approximated as function of temperature and pressure as follows:
$$
\eta_{F}=\left(\frac{i^{2}}{f_{11}+f_{12} \cdot T+\dot{i}^{2}}\right) \cdot\left(f_{21}+f_{22} \cdot T\right)
$$

```python
#%% FARADAY EFFICIENCY

eta_farad = (i**2 /(f_11 + f_12 * T + i**2)) * (f_21 + f_22* T)
print('The faradaic efficiency is ' + str(round(eta_farad, 2)))
```

## Power curve
The power required to drive the whole process, is the main KPI together with the energy production. Power is defined as:
$$
W_{\text {stack }}=V_{\text {stack }} \cdot I=\left(V_{\text {cell }} \cdot N\right) \cdot\left(i \cdot A_{\text {cell }}\right)
$$

```python
P_cell  = V_cell * i * A
P_stack = P_cell * N /1000 # Power in kW
```

## Diffusion to oxigen layer
In modern Alkaline electrolyzer cells it can occur the phenomena of having moles of Hydrogen diffusing to the Oxigen electrode, and this is likely to being avoided, in order to minimize Hydrogen losses. It is approximated as:
$$
\mathrm{HTO}=\left[C_{1}+C_{2} \cdot \mathrm{T}+\mathrm{C}_{3} \cdot \mathrm{T}^{2}+\left(\mathrm{C}_{4}+\mathrm{C}_{5} \cdot \mathrm{T}+\mathrm{C}_{6} \cdot \mathrm{T}^{2}\right) \cdot \exp \left(\frac{\mathrm{C}_{7}+\mathrm{C}_{8} \cdot \mathrm{T}+\mathrm{C}_{9} \cdot \mathrm{T}^{2}}{\mathrm{i}}\right)\right. \\
\times]+\left[\mathrm{E}_{1}+\mathrm{E}_{2} \cdot p+\mathrm{E}_{3} \cdot p^{2}+\left(\mathrm{E}_{4}+\mathrm{E}_{5} \cdot p+\mathrm{E}_{6} \cdot p^{2}\right) \cdot \exp \left(\frac{\mathrm{E}_{7}+\mathrm{E}_{8} \cdot p+\mathrm{E}_{9} \cdot p^{2}}{i}\right)\right]
$$

```python
HTO = (C_1 + C_2*T + C_3*T**2 + (C_4 + C_5*T + C_6*T**2)*np.exp((C_7 + C_8*T + C_9*T**2)/i))\
    + (E_1 + E_2 * p + E_3 * p **2 + (E_4 + E_5 * p + E_6 * p **2)*np.exp((E_7 + E_8 * p + E_9 * p **2)/i))
```

## Mass balance

The overall reaction is:
$$
\mathrm{H}_{2} \mathrm{O} \rightarrow \mathrm{H}_{2}+\frac{1}{2} \mathrm{O}_{2}
$$

The mass produced via electrolysis can be easily  calculated by mean of Farady's law:
$$
n_{\mathrm{H}_{2}, \text{prod}}=\eta_{\text{Faraday}} \cdot \frac{I \cdot t}{{z} \cdot {F}} \cdot {N}=\eta_{\text{Faraday}} \cdot \frac{i \cdot A \cdot t}{{z} \cdot {F}} \cdot {N}
$$
where:
* $I$ is the current passed through the electrodes in a given time interval $t$ = 1 hour.
* $i$ is the current density 
* $F$ is the Faraday constant 96485 C/mol.
* $z$ is the number of electrons involved in the redox reaction: for water splitting process is 2.

The individual products production are avaluated as:
where:
* $n_{\mathrm{H}_{2}, \text { cat }}=n_{\mathrm{H}_{2}, \text { prod }}$ is the actual hydrogen production at the cathode
* $n_{\mathrm{H}_{2}, \text { an }}=n_{\mathrm{HTO}}$ is the  hydrogen diffused to the oxigen eletrode (anode)
* $n_{\mathrm{O}_{2}, \text { an }}=n_{\mathrm{O}_{2}, \text { prod }}=\frac{1}{2} n_{\mathrm{H}_{2}, \text { prod }}$ is the Oxygen produced at the anode, calculated with the stoichiometric ratio from Hydrogen at the cathode
* $n_{\mathrm{H}_{2} \mathrm{O}}=n_{\mathrm{H}_{2}, \text { prod }}$ is the number of moles required to power the process


```python
H2_prod_cell   = eta_farad * i*A/(n*F)            # mol/sec
H2_prod_stack  = H2_prod_cell*N*3.6* (2*H.mass)   # kg/hour 
```

## Energy balance
The thermoneutral voltage is calculated as the equilibrium voltage in which the electrolyzer us balanced and no heat is produced or generated, defined as:
$$
V_{t n}=\frac{\Delta H}{Z \cdot F}
$$

Since the voltage of the cell was calculated previously, the generated heat can be indentified as:
$$
Q_{g e n}=N \cdot I \cdot\left(V_{\text {cell }}-V_{t n}\right)
$$

```python
heat_develop_cell = (V_cell - V_tn) * i * A
heat_develop_stack = heat_develop_cell * N /1000 # kW produced from the whole stack
```

# Efficiency
The system efficiency can be intended as  the ratio between "what you get and what you pay". The ratio is therefore calculated as the energy content of the output stream, over the energy input:
$$\eta_{en}=\frac{m_{\text{H2}, \mathrm{OUT}} \cdot \mathrm{LHV}_{\text{H2}}}{{W}_{input}}$$
where:
* ${W}_{input}$ is defined by the components embedded in the system
* Additionally, LHV$_{H_{2}}$ is 120 MJ/kg
* m$_{H_{2}}$ is 17.7 kg/hr for a 1MW AEC electrolyser
* System efficiency is 59\%.

At the current setpoint the efficiency is evaluated as:
$$\eta_{en}=\frac{17.7{\text{kg H2/hr}} \cdot 120 \text{ MJ/kg}}{1.024 \text{ MW} \cdot 3600 \text{ s}} = 58\%$$

```python
i_temp = np.linspace(0,5000,10000)
LHV_H2 = 120 # MJ/kg

# CELL VOLTAGE
V_cell_plot = V_rev + ((r_1 + d_1) + r_2 * T + d_2 * p) * i_temp + s * np.log10((t_1 + t_2/T + t_3 / T**2)*i_temp + 1)
p_plot = V_cell_plot* i_temp * A * N /1000

# FARADAIC EFFICIENCY
eta_farad = (i_temp**2 /(f_11 + f_12 * T + i_temp**2)) * (f_21 + f_22* T)
#eta_farad = 0.9

# MASS 
H2_prod_stack_temp  = eta_farad * i_temp*A/(n*F)*N*3.6* (2*H.mass)   # kg/h

# EFFICIENCY
efficiency = H2_prod_stack_temp *LHV_H2/(3600*p_plot/1000) * 100

# HEAT CALCULATION
heat_dev_cell_plot = (V_cell_plot - V_tn) * i_temp * A
heat_stack_plot = heat_dev_cell_plot * N /1000 # kW produced from the whole stack

# VOLTAGE - POWER PLOT
color = 'tab:blue'
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(i_temp, V_cell_plot, label = 'Voltage')
ax.set_xlabel('Current [A]')
ax.set_ylabel('Voltage [V]', color = color)

color = 'tab:orange'
ax2 = ax.twinx()
ax2.plot(i_temp, p_plot, color = color, label = 'Power')
ax2.set_ylabel('Power produced [kW]', color = color)
ax.margins(x=0, y=0)
ax2.margins(x=0, y=0)
fig.legend()
plt.show()


# EFFICIENCY PLOT

color = 'tab:blue'
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(p_plot, efficiency)

ax.set_xlabel('Load [kW]')
ax.set_ylabel('System efficiency [%]', color = color)
plt.axvline(p_plot.max() *0.16, linestyle = 'dashed', color = 'tab:red', label = 'Unfeasible area [16% Max Load]')
plt.axvspan(0, p_plot.max() *0.16, ymin=0.0, ymax=1, alpha=0.2, color='red')

color = 'tab:orange'
ax2 = ax.twinx()
ax2.plot(p_plot, H2_prod_stack_temp, color = color)
ax2.set_ylabel('Hydrogen production [kg/h]', color = color)
ax.margins(x=0, y=0)
ax2.margins(x=0, y=0)
fig.legend()
plt.show()


```
